```python
#     3)ALGORITHMS,TRAINING AND PREDICTION
```

```python
#trai test spilt method

from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
#dision tree cassifier
print("using decision tree algorithm")
c=DecisionTreeClassifier(max_depth=4)
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
print("predicted value=",y_pred)

#accuracy
acc=metrics.accuracy_score(y_test,y_pred)*100
print("accuracy=",acc)

#precision
prs=precision_score(y_test,y_pred,pos_label=1)
print("precision=",prs)

#recall
rcl=recall_score(y_test,y_pred,pos_label=1)
print("recall=",rcl)

#f1-score
f1s=f1_score(y_test,y_pred,pos_label=1)
print("f1-scare=",f1s)
```

```python
#confusion matrics for decision tree
from sklearn.metrics import confusion_matrix
label=['valid','fraud']
cmt=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(10,10))
sb.heatmap(cmt,xticklabels=label,yticklabels=label,annot=True,fmt="d");
plt.title("confusion matrix using random forest")
plt.ylabel('True class')
plt.xlabel('predicted class')
plt.show()
```

```python
#applying random forest algorithm
print("using random forest algorithm")
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred1=rfc.predict(x_test)
print("predicted value=",y_pred1)

#accuracy
acc1=accuracy_score(y_test,y_pred1)
print("accuracy=",acc1)

#precision
prs1=precision_score(y_test,y_pred1)
print("precision=",prs1)

#recall
rcl1=recall_score(y_test,y_pred1)
print("recall=",rcl1)

#f1-score
f1s1=f1_score(y_test,y_pred1)
print("f1-score=",f1s1)
```

```python
#confusion matrix for random forest
labels=['valid','fraud']
cmt1=confusion_matrix(y_test,y_pred1)
plt.figure(figsize=(10,10))
sb.heatmap(cmt1,xticklabels=label,yticklabels=label,annot=True,fmt="d");
plt.title("confusion matrix using random forest")
plt.ylabel('True class')
plt.xlabel('predicted class')
plt.show()
```
